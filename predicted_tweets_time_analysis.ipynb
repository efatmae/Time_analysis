```python
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
```

```python
predicted_tweets = pd.read_csv("Data/turkish_tweets_labels_predicted_20_08_2018.csv")
```

```python
predicted_tweets.columns
```

```python
predicted_tweets = predicted_tweets[['id', 'created_at', 'text', 'prediction', 'non_vio_prob',
       'vio_prob']]
```

```python
predicted_tweets["created_at_dt"] = pd.to_datetime(predicted_tweets["created_at"])
```

```python
predicted_tweets_dt = predicted_tweets.set_index(predicted_tweets["created_at_dt"])
```

```python
predicted_tweets_dt["created_at_dt"].describe()
```

```python
len(predicted_tweets_dt.resample("D"))
```

```python
len(predicted_tweets_dt.resample("H"))
```

```python
plt.figure(figsize=(20,4))

plt.plot(predicted_tweets_dt['prediction'].resample("D", how='count'))
plt.show()
```

```python


plt.figure(figsize=(20,4))

plt.plot(predicted_tweets_dt[predicted_tweets_dt['prediction']==1].resample("D", how='count'))
plt.show()
```

```python
plt.figure(figsize=(15,4))

plt.plot(predicted_tweets_dt[predicted_tweets_dt['prediction']==1].resample("H", how='count'))
plt.show()
```
